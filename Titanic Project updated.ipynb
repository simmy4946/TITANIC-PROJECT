```python
# check for null values
df.isnull().sum().any()
```

```python
# check null values in the data , there is null values are presented , and it will done by required treatment...
df.isnull().sum()
```

```python
# Drop the null values
df.dropna(axis=0, inplace=True)
df.head()
```

```python
# Fill null values..
df['Age']= df['Age'].fillna(df['Age'].mean())
```

```python
# check again if there is any left , now treated well , and no more null values..
df.isnull().sum()
```

```python
# Make a countplot of gender...
ax = sns.countplot(x= 'Sex', data=df,color='m')

for bars in ax.containers:
    ax.bar_label(bars)
```

This above graph is showing number of male and female presented,male are 95 and female are 88, this means men are more than female in tiatanic.

```python
# Now, we are doing groupby of sex with number of people (male & female) survived..
df.groupby(['Sex'], as_index=False)['Survived'].sum().sort_values(by='Survived', ascending=False)
```

```python
# Make a countplot of sex and no of people survived...
ax = sns.countplot(data= df, x= 'Survived', hue= "Sex",color='yellow')

for bars in ax.containers:
  ax.bar_label(bars)
```

This above graph is showing number of people survived whether it is male or female, there 82 female and 41 males survived , however 6 female and 54 male not survived.Female are survived more as compare to men while total number of male are greater than women, but though women survived more.

```python
# Make a countplot of Pclass with sex..
ax = sns.countplot(data= df, x= 'Pclass', hue= "Sex",color='aqua')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Groupby of of sex and class
df.groupby(['Sex'], as_index=False)['Pclass'].sum().sort_values(by='Pclass', ascending=False)
```

From this above graph, we getting information about the all male and female in the pclass, male are prefer pclass 1 as compare to the female. Female are mostly using pclass 2 and 3.Men are lesser in pclass 2 and 3.

```python
# Make a countplot of pclass and number of people survived..
ax = sns.countplot(data= df, x= 'Pclass', hue= "Survived",color='red')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# By doing groupby , we getting an information about pclass and number of people survived in pclasses..
df.groupby(['Pclass'], as_index=False)['Survived'].sum().sort_values(by='Survived', ascending=False)
```

From this above graph, we are knowing about the survived people in various classes, 106 people are survived in pclass1 , 12 people survived in pclass2 and 5 people survived in pclass3.It shows people are more survived in pclass 1 rather than pclass2 and 3.
